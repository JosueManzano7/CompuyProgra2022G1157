#PROBLEMA 3.1
Escriba un programa tal que dado como datos 270 números enteros, obtenga la suma de los números impares y el promedio de los números pares.

```python
import random
e=[random.randint(0,1000) for x in range(270)]#generador de numeroes enteros
SUMPAR=0
SUMIMP=0
CUEPAR=0
I=1
while I<=270:
    NUM=e[I-1]
    print(f"Valor {I}: {e[I-1]}")
    if NUM!=0:
        if NUM%2==0:
            SUMPAR+=NUM
            CUEPAR+=1
        else:
            SUMIMP+=NUM
    I+=1
PROPAR=SUMPAR/CUEPAR
print("PROMEDIO DE NUMEROS PARES: ", round(PROPAR,2))
print("SUMA DE NUMEROS IMPARES: ", SUMIMP)
```

#PROGAMA 3.3
Escriba un programa que lea un número entero N y calcule el resultado de la siguiente serie:

1-(1/2)+(1/3)-(1/4)+...+(1/N)

```python
SERIE=0
N=int(input("Dame el número de términos de la serie: "))
BAND="T"
I=1
while(I<=N):
    if BAND=="T":
        SERIE=SERIE+(1/I)
        BAND="F"
    else:
        SERIE=SERIE-(1/I)
        BAND="T"
    I+=1
print("RESULTADO CON ",N," TERMINOS: ",round(SERIE,4))
    
```

#PROBLEMA 3.4
Calcule el aumento de sueldo para un grupo de empleados de una empresa teniendo en cuenta el siguiente criterio:

- Si el sueldo es inferior a \$1000 : Aumento 15%
- Si el sueldo es mayor o igual a \$1000 : Aumento 12%

Imprima el sueldo nuevo del trabajador y el total de nómina de la empresa, considerando este nuevo aumento.


```python
NOM=0
SUE=float(input("Dame el sueldo del trabajador 1: "))
i=1
while SUE!=-1:
    if SUE<1000:
        NSUE=SUE*1.15
    else:
        NSUE=SUE*1.12
    NOM=NOM+NSUE
    print("NUEVO SUELDO: ",round(NSUE,2))
    SUE=float(input(f"Dame el sueldo del trabajador {i+1}: "))
    i+=1

print("NOMINA: ",NOM)
```
